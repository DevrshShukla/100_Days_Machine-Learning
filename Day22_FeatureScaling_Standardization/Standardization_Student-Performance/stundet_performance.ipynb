## Student Performance Data

```python
import pandas as pd

data = {
    "Study_Hours": [2, 5, 8, 1, 6, 9, 4, 7, 3, 10],
    "Attendance": [65, 80, 90, 55, 85, 95, 70, 88, 60, 98],
    "Prev_Score": [45, 60, 85, 35, 70, 92, 50, 78, 48, 95],
    "Family_Income": [30000, 50000, 120000, 25000, 70000, 150000, 45000, 90000, 35000, 200000],
    "Final_Score": [42, 62, 88, 38, 72, 94, 55, 80, 46, 97]
}

df = pd.DataFrame(data)
print(df)
```

```python
x = df.drop("Final_Score", axis=1)
y = df['Final_Score']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

### Applying Standardization 

```python
scaler = StandardScaler()
x_train_nostd = scaler.fit_transform(x_train)
x_test_nostd = scaler.transform(x_test)
```

```python
x_train_nostd = pd.DataFrame( x_train_nostd , columns= x_train.columns)
x_test_nostd = pd.DataFrame( x_test_nostd , columns= x_test.columns)    
```
