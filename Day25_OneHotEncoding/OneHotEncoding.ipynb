### One Hot Encoding - > use for output columns

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cars.csv')
df.sample(5)
```

```python
df['fuel'].value_counts()
```

## One Hot Encoding using Pandas

```python
pd.get_dummies( df , columns=['fuel' , 'owner'])
```

## K-1 One Hot Encoding
**drop first column**

```python
pd.get_dummies( df , columns=['fuel' , 'owner'] , drop_first=True)
```

##  One Hot Encoding using sklearn

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(df.iloc[ : , :4] , df.iloc[ : , -1] , test_size = 0.2)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder( drop='first' , sparse_output=False , dtype=np.int32 )
```

```python
x_train_n = ohe.fit_transform( x_train[ [ 'fuel' , 'owner' ] ] )
x_test_n = ohe.transform( x_train[ [ 'fuel' , 'owner' ] ] )
```

### One Hot Encoding with top categories

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].unique()
threshold = 100
```

```python
replace = counts[ counts <= threshold ].index
replace
```

```python
pd.get_dummies( df['brand'].replace( replace , 'Other' ) , drop_first=True )
```
