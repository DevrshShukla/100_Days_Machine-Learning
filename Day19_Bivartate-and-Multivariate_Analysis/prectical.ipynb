```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### load datasets

```python
titanic = pd.read_csv('train.csv')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

## 1. Scatter Plot 
**[ Numerical - Numerical ]**

### Bivariated analysis

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

### Multivariated analysis


```python
sns.scatterplot(x='total_bill' , y='tip' , hue='sex' , data=tips)
```

### 2. Bar plot
**[ Numerical - Categorical ]**

### Bivariated analysis

```python
sns.barplot( x='Pclass' , y='Age' , data=titanic)
```

### Multivariated analysis

```python
sns.barplot( x='Pclass' , y='Age' , hue='Sex' ,  data=titanic)
```

### 3. Box Plot
**[ Numerical - Categrical ]**

```python
sns.boxplot( x='day' , y='total_bill', data=tips)
```

### 4. Dist Plot
**[ Numerical - categrical ]**

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'] , hist=False , color='red' , label='Not Survived')
sns.distplot(titanic[titanic['Survived'] == 1]['Age'] , hist=False , color='blue' , label='Not Survived')
```

### 5. Heat Map
**[ Categorical - Categorical ]**

```python
x = pd.crosstab(titanic['Survived'] , titanic['Pclass'])
sns.heatmap(x , annot=True , fmt='d')
```

### 6. Cluster Map
**[ Categorical - Categorical ]**

```python
sns.clustermap(x , annot=True , fmt='d')
```

### 7. Pair Plot
**[ Numerical - Numerical ]**

```python
sns.pairplot(iris , hue='species')
```
