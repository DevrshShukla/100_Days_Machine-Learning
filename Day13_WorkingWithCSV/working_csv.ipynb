## index col parameter

### header parameter

```python
pd.read_csv('person.csv')
```

### use_cols parameter

```python
pd.read_csv('person.csv' , usecols=['Age' , 'Salary'] )
```

### skip rows

```python
pd.read_csv('person.csv' , nrows=15 )
pd.read_csv('person.csv' , skiprows=5 )
```

### skip bad lines

```python
# skip lines that accure error
pd.read_csv('person.csv' , sep=';' , error_bad_lines=False) 
```

### dtypes parameter

```python
#change the datatype of a column

# pd.read_csv('person.csv').info()
pd.read_csv('person.csv' , dtype={'Age':float} ).info()
```

### Handling Dates

```python
# pass data in string -> convert to data formate

pd.read_csv(' file_name' , parse_dates=['DOB'] )
```

### converters

```python
# Mumbai indians - MI

def convert_team_name(name):
    if name == 'MI':
        return 'Mumbai Indians'
    elif name == 'RCB':
        return 'Royal Challengers Bangalore'
    else:
        return name
    
pd.read_csv('matches.csv' , converters={'team1':convert_team_name , 'team2':convert_team_name} )
```

### na_values parameter

```python
# values = NaN

pd.read_csv('person.csv' , na_values=['Not Available' , 'n.a.' , 'NA' , '--'] )
```

### load huge dataset in chunks

```python
df = pd.read_csv('person.csv' , chunksize=5)
for chunk in df:
    print(chunk)
```
